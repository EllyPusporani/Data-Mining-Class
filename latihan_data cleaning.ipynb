# Imputation Data

```python
#number of NA for each variable
np.sum(data.isnull())
```

```python
# mendeteksi missing value (NA)
data.isnull().sum()
```

```python
# total data NA di seluruh data
data.isnull().sum().sum()
```

```python
#imputasi missing value data numerik [mean / median]
data['X1'] = data['X1'].fillna((data['X1'].mean())) 
data['X2'] = data['X2'].fillna((data['X2'].mean()))
data['X3'] = data['X3'].fillna((data['X3'].mean())) 
data['X4'] = data['X4'].fillna((data['X4'].mean()))
data['X5'] = data['X5'].fillna((data['X5'].median()))
data.isnull().any()
```

```python
#imputasi data kategorik dengan [modus]
data['Y'] = data.fillna(data['Y'].value_counts().index[0])
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

# Encode Label

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# membuat koding [Encode] untuk Y tanpa membuat kolom baru
data['Y1'] = LabelEncoder().fit_transform(data['Y1']) 
data['Y'] = LabelEncoder().fit_transform(data['Y'])
data.head()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

# Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['X1'] = StandardScaler().fit(data[['X1']]).transform(data[['X1']])
data['X2'] = StandardScaler().fit(data[['X2']]).transform(data[['X2']])
data['X3'] = StandardScaler().fit(data[['X3']]).transform(data[['X3']])
data['X4'] = StandardScaler().fit(data[['X4']]).transform(data[['X4']])
data['X5'] = StandardScaler().fit(data[['X5']]).transform(data[['X5']])
data.head()
```

# Normalize

```python
from sklearn.preprocessing import Normalizer
```

```python
data['X1'] = Normalizer().fit(data[['X1']]).transform(data[['X1']])
data['X2'] = Normalizer().fit(data[['X2']]).transform(data[['X2']])
data['X3'] = Normalizer().fit(data[['X3']]).transform(data[['X3']])
data['X4'] = Normalizer().fit(data[['X4']]).transform(data[['X4']])
data['X5'] = Normalizer().fit(data[['X5']]).transform(data[['X5']])
data.head()
```
